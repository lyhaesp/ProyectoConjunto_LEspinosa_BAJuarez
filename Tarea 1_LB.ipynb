### Ejercicio 3: Programa que muestre los números primos del 0-100. (Estudiante 2)

```python
def num_primo(fin):
    num = 1
    while num <= fin:
        cont = 1
        x = 0
        while cont <= num:
            if num % cont == 0:
                x = x+1
            cont = cont + 1
        if x==2:
            print(num)
        num = num + 1
num_primo(100)
```
